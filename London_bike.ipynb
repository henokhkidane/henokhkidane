```python
# building a dictionary to rename the columns to my prefered name

new_name_dic = {
    "timestamp" : "time" ,
    "cnt" : "count" ,
    "t1" : "real_temp_C" ,
    "t2" : "feelslike_temp_C",
    "hum" : "humidity" ,
    "wind_speed" : "wind_speed" ,
    "weather_code" : "weather" ,
    "is_holiday" : "is_holiday" ,
    "is_weekend" : "is_weekend" ,
    "season" : "season"
}

df.rename(new_name_dic, axis=1, inplace = True )
```

```python
df["weather"].value_counts()
```

```python
# building another 2 dictionary for weather and season
weather_dictionary = {
    "1.0" : "Clear" ,
    "2.0" : "Scattered cloud" ,
    "3.0" : "Broken Cloud" ,
    "4.0" : "Cloudy" ,
    "7.0" : "Rain" ,
    "10.0": "Rain with tounderstorm" ,
    "26.0": "Snowfall"
}
season_dictionary = {
    "0.0" : "Spring" ,
    "1.0" : "Summer" ,
    "2.0" : "Autumen",
    "3.0" : "Winter"

}
```

```python
 # using the map function to change the values of weather and season with the new data in the dictionary
 df["weather"] = df["weather"].astype("str")
 df["weather"] = df["weather"].map(weather_dictionary)
 df["season"] = df["season"].astype("str")
 df["season"] = df["season"].map(season_dictionary)
```
